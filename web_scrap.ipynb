```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
#we are sending request to  the url and taking response
url = 'https://www.playstation.com/en-us/explore/ps4/'
response=requests.get(url)
response    #here the 200 is the status of success
```

```python
#for retriving information
response.text
```

```python
#now here generally response is the retrivedalong with some tags of xml for saperate that xml-tags
soup=BeautifulSoup(response.text , 'lxml')
soup
```

```python
#for searching anchor-tag in soup:
for Alink in soup.find_all('a'):
    print(Alink)  #print anchor-tags
    print(Alink.get('href'))  #for printing href (as 'href'as it consiting of '=') are generally comes under the anchor-tags
    
```

```python
#for retriving all content(pure) like paragraph-text
for ptag in soup.find_all('p'):
    print(ptag)
#now  second loop for b tag only text-part of it:
    for btag in ptag.find_all('b'):
        print(btag.text)
```
