```python
import matplotlib.pyplot as plt
```

```python
x=[2,3]
y=[9,5]
a=[1,10]
b=[5,3]
```

```python
plt.xlabel("speed")   #giving name to the x-axes
plt.ylabel("time")    #giving name to the y-axes
plt.plot(x,y,label='water',color='blue')  #alloting parameters of x & y
plt.plot(a,b,label='soil',color='red')   #alloting parameters of a & b
plt.bar(x,y)    #for plotting bar-graph of x & y
plt.bar(a,b)    #for plotting bar-graph of a & b 
plt.legend()    #for showing giving parameters
plt.grid(color='green')   #faciliating grids along with color
plt.xlim(0,5)     #limits of x
plt.ylim(0,10)    #limits of y
plt.title("title here..")   #for giving title-name of whole plotted-visualization
plt.show()  #finally for showing plotted window or to visualize
```
